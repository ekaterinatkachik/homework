```python
def pros(n, k=2):
    if not isinstance(n, int):
        raise TypeError("Должно быть число")
    if n < 1:
        raise ValueError("Число должно быть натуральным")
    if n == 1:
        return []
    if n % k == 0:
        return [k] + pros(n // k, k)
    else:
        return pros(n, k + 1)
```

```python
%%unittest_main


class TestSmallNumber(unittest.TestCase):
    def test_pros(self):
        self.assertEqual(pros(6), [2, 3], "should be [2, 3]")


class TestProsNumber(unittest.TestCase):
    def test_pros(self):
        self.assertEqual(pros(17), [17], "should be [17]")


class TestOne(unittest.TestCase):
    def test_pros(self):
        self.assertEqual(pros(1), [], "should be []")
```

```python
import numpy as np
x = list(map(float, input().split()))
y = list(map(float, input().split()))


def coeff(x, y):
    x = np.array(x)
    y = np.array(y)
    n = len(x)
    sumx = 0
    sumx2 = 0
    sumy = 0
    sumxy = 0
    for i in range(n):
        sumx += x[i]
        sumx2 += (x[i])**2
        sumy += y[i]
        sumxy += x[i]*y[i]
    a = (n*sumxy-sumx*sumy)/(n*sumx2-sumx**2)
    return a


print(coeff(x, y))
```

```python
%%unittest_main


class TestPerfect(unittest.TestCase):
    def test_coeff(self):
        x = [1, 2]
        y = [3, 6]
        result = coeff(x, y)
        self.assertEqual(result, 3.0)


class TestZeroSlope(unittest.TestCase):
    def test_coeff(self):
        x = [1, 2, 3]
        y = [5, 5, 5]
        result = coeff(x, y)
        self.assertEqual(result, 0.0)


class TestMinimalPoints(unittest.TestCase):
    def test_coeff(self):
        x = [1, 0]
        y = [0, 1]
        result = coeff(x, y)
        self.assertEqual(result, -1.0)
```

```python
import random


def quick_sort(a):
    assert isinstance(a, list), "На входе должен быть список"
    if len(a) > 0:
        for elem in a:
            assert isinstance(elem, (int, float)), "Все элементы должны быть числами"
    if len(a) <= 1:
        return a
    else:
        q = random.choice(a)
        l = []
        m = []
        r = []
        for elem in a:
            if elem < q:
                l.append(elem)
            elif elem > q:
                r.append(elem)
            else:
                m.append(elem)

        left_sorted = quick_sort(l)
        right_sorted = quick_sort(r)

        result = left_sorted + m + right_sorted
        assert isinstance(result, list), "Результатом должен быть список"

        return result


print(quick_sort([1, 2, 5, 4, 'a']))
```

```python
class Caesar:
    alphabet = "яюэьыъщшчцхфутсрпонмлкйизжёедгвба"

    def __init__(self, key):
        if isinstance(key, str):
            try:
                key = int(key)
            except ValueError:
                assert False, "Ключ должен быть целым числом"
        elif not isinstance(key, int):
            assert False, "Ключ должен быть целым числом"

        assert len(self.alphabet) > 0, "Алфавит пустой"
        assert len(self.alphabet) == len(set(self.alphabet)), "В алфавите есть повторяющиеся буквы"

        self._encode = dict()
        self._decode = dict()
        n = len(self.alphabet)
        for i in range(n):
            letter = self.alphabet[i]
            encoded = self.alphabet[(i + key) % n]
            self._encode[letter] = encoded
            self._encode[letter.upper()] = encoded.upper()
            self._decode[encoded] = letter
            self._decode[encoded.upper()] = letter.upper()

        for char in self.alphabet:
            enc_char = self._encode[char]
            dec_char = self._decode[enc_char]
            assert dec_char == char, f"Нет обратимости у символа '{char}'"

    def encode(self, text):
        assert isinstance(text, str), "Входной текст должен быть строкой"
        return ''.join([self._encode.get(char, char) for char in text])

    def decode(self, line):
        assert isinstance(line, str), "Входная строка должна быть строкой"
        return ''.join([self._decode.get(char, char) for char in line])


raw_key = input('Введите ключ: ')
cipher = Caesar(raw_key)
line = input()
while line:
    print(cipher.encode(line))
    line = input()
```
