```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
fig = plt.figure(figsize = (10,5)) 
ax1 = fig.add_subplot(111) 

x_measured = [60.6,77,98.4,111.8,137]
y_measured = [315,395,505,575,705]

x = [58, 140]
y = np.interp(x, x_measured, y_measured)

ax1.scatter(x_measured, y_measured, marker='.', color='k')

ax1.errorbar(x_measured, y_measured, yerr=10, xerr = 2, color = 'k', linestyle = 'None')

ax1.plot(x,y, 'r')

ax1.grid() 

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('D:\info\iris_data.csv')
a = list(df['Species'])
setosa = a.count('Iris-setosa')
versicolor = a.count('Iris-versicolor')
virginica = a.count('Iris-virginica')
n = setosa+versicolor+virginica
plt.pie([setosa/n, versicolor/n, virginica/n], labels = ['Iris-setosa','Iris-versicolor','Iris-virginica'])
plt.title('The proportion of different types of iris')
plt.show()
b = list(df['PetalLengthCm'])
k1 = 0
k2 = 0
k3 = 0
for i in range(len(b)):
    if b[i] < 1.2:
        k1 += 1
    elif b[i] > 1.5:
        k3 += 1
    else:
        k2 += 1
plt.pie([k1/n, k2/n, k3/n], labels = ['Less than 1.2 cm','More than 1.2 cm and less than 1.5 cm','More than 1.5 cm'])
plt.title('Petal length')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.subplot(211)
plt.title('BTC')
plt.xlabel('close')
plt.ylabel('date')
df = pd.read_csv('D:\info\BTC_data.csv')
date = list(df['time'])
for i in range(len(date)):
    date[i] = date[i][:10]
close = list(df['close']) 
plt.plot(date, close)

plt.subplot(212)
x = np.array([j + i for j in range(len(date))])
y = np.array(close)
z = np.polyfit(x, y, 10)
p = np.poly1d(z)
plt.plot(z, p)
```
