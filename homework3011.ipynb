# Задача 1

```python
n = int(input())
a = list(map(int, input().split()))
k = int(input())


def build_tree(a):
    tree = {}
    for i in range(n):
        tree[n + i] = a[i]
    for i in range(n - 1, 0, -1):
        tree[i] = max(tree[2 * i], tree[2 * i + 1])
    return tree


def max_on_segment(tree, l, r):
    _max = 0
    l += n
    r += n
    while l < r:
        if l & 1:
            _max = max(_max, tree[l])
            l += 1
        if r & 1:
            r -= 1
            _max = max(_max, tree[r])
        l //= 2
        r //= 2
    return _max


for i in range(k):
    l, r = map(int, input().split())
    tree = build_tree(a)
    res = max_on_segment(tree, l-1, r)
    print(res)
```

# Задача 2


```python
n = int(input())
a = list(map(int, input().split()))
k = int(input())


def gcd(a, b):
    while b:
        a, b = b, a % b
    return a


def build_tree(a):
    tree = {}
    for i in range(n):
        tree[n + i] = a[i]
    for i in range(n - 1, 0, -1):
        tree[i] = gcd(tree[2 * i], tree[2 * i + 1])
    return tree


def gcd_on_segment(tree, l, r):
    res = 0
    l += n
    r += n
    while l < r:
        if l & 1:
            res = gcd(res, tree[l])
            l += 1
        if r & 1:
            r -= 1
            res = gcd(res, tree[r])
        l //= 2
        r //= 2
    return res


for i in range(k):
    l, r = map(int, input().split())
    tree = build_tree(a)
    res = gcd_on_segment(tree, l-1, r)
    print(res)
```
